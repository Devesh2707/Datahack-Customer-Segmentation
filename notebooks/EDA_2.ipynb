```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('fivethirtyeight')
```

```python
train = pd.read_csv("/run/media/devesh/Hard Disk/Projects/Datahack-Customer-Segmentation/input/Train.csv")
test = pd.read_csv("/run/media/devesh/Hard Disk/Projects/Datahack-Customer-Segmentation/input/Test.csv")
```

```python
df = train.copy()
```

```python
df = df.sort_values('ID').reset_index(drop = True)
```

```python
set_id_train = set(df['ID'])
set_id_test = set(test['ID'])
```

```python
common_ids = list(set_id_train.intersection(set_id_test))
```

```python
common_ids = np.array(common_ids).astype(np.int32)
```

```python
common_ids = np.sort(common_ids).tolist()
```

```python
df_test = test.copy()
```

```python
# df_test = df_test.set_index('ID')
```

```python
common_df = pd.merge(df, df_test, on = 'ID')
```

```python
common_df.head()
```

```python
common_seg = pd.DataFrame()

common_seg['ID'] = common_df['ID']
common_seg['Segmentation'] = common_df['Segmentation']
```

```python
common_seg.head()
```

```python
df_test = df_test.set_index('ID')
```

```python
uncommon_test = df_test.drop(common_ids, axis = 0)
```

```python
uncommon_test
```

```python
common_test = pd.merge(test, common_seg, on = 'ID')
```

```python
new_train = pd.concat([train, common_test]).sort_values("ID").reset_index(drop = True)
```

```python
len(new_train)
```

```python
new_train['Work_Experience'] = new_train['Work_Experience'].fillna(1.0)
new_train['Profession'] = new_train['Profession'].fillna('Artist')
new_train['Graduated'] = new_train['Graduated'].fillna('No')
new_train['Ever_Married'] = new_train['Ever_Married'].fillna('No')
```

```python
new_train['Family_Size'] = new_train['Family_Size'].fillna(2)
new_train['Var_1'] = new_train['Var_1'].fillna('Cat_6')
```

```python
new_train.isnull().sum()
```

```python
target = new_train.pop('Segmentation')
```

```python
from sklearn.preprocessing import LabelEncoder

encoder_target = LabelEncoder()

target = encoder_target.fit_transform(target)
```

```python
cat_cols = new_train.select_dtypes(include = 'object').columns
```

```python
cat_cols
```

```python
encoder_data = LabelEncoder()

for col in cat_cols:
    new_train[col] = encoder_data.fit_transform(new_train[col])
```

```python
new_train.head()
```

```python
new_train.drop('ID', axis = 1, inplace = True)
```

```python
from sklearn.model_selection import cross_val_score
from lightgbm import LGBMClassifier

model = LGBMClassifier()

scores = cross_val_score(model, new_train, target, cv = 5, scoring = 'accuracy')
```

```python
scores.mean()
```

```python
uncommon_test['Work_Experience'] = uncommon_test['Work_Experience'].fillna(1.0)
uncommon_test['Profession'] = uncommon_test['Profession'].fillna('Artist')
uncommon_test['Graduated'] = uncommon_test['Graduated'].fillna('No')
uncommon_test['Ever_Married'] = uncommon_test['Ever_Married'].fillna('No')
uncommon_test['Family_Size'] = uncommon_test['Family_Size'].fillna(2)
uncommon_test['Var_1'] = uncommon_test['Var_1'].fillna('Cat_6')
```

```python
for col in cat_cols:
    uncommon_test[col] = encoder_data.fit_transform(uncommon_test[col])
```

```python
model.fit(new_train, target)
```

```python
preds = model.predict(uncommon_test)
```

```python
uncommon_test.head()
```

```python
uncommon_seg = pd.DataFrame()

uncommon_seg['ID'] = list(uncommon_test.index)
uncommon_seg['Segmentation'] = preds
```

```python
final = pd.concat([common_seg, uncommon_seg]).reset_index(drop = True)
```
